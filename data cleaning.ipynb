```python
data162 = data16[['What is your age?', 'What is your gender?',
                  'What country do you work in?', 
                  'If you have a mental health issue, do you feel that it interferes with your work when NOT being treated effectively?',
                  'Do you work remotely?', 
                  'Does your employer provide mental health benefits as part of healthcare coverage?',
                  'Has your employer ever formally discussed mental health (for example, as part of a wellness campaign or other official communication)?',
                  'Do you think that discussing a mental health disorder with your employer would have negative consequences?',
                  'Would you feel comfortable discussing a mental health disorder with your coworkers?',
                  'Would you bring up a mental health issue with a potential employer in an interview?',
                  'Have you been diagnosed with a mental health condition by a medical professional?',
                  'If so, what condition(s) were you diagnosed with?']]


data162 = data162.rename({'What is your age?': 'Age', 'What is your gender?': 'Gender', 
                          'What country do you work in?': 'Country',
                         'Does your employer provide mental health benefits as part of healthcare coverage?':
                         'Does your employer provide mental health benefits?',
                         'Do you think that discussing a mental health disorder with your employer would have negative consequences?':
                         'Do you think that discussing a mental health issue with your employer would have negative consequences?',
                          'Has your employer ever formally discussed mental health (for example, as part of a wellness campaign or other official communication)?':
                          'Would you be willing to discuss a mental health issue with your direct supervisor(s)?',}, axis=1)
data162['year'] = 2016
data162
```

```python
data142 = data14[['Age', 'Gender', 'Country', 'If you have a mental health condition, do you feel that it interferes with your work?',
                  'Do you work remotely (outside of an office) at least 50% of the time?', 
                  'Does your employer provide mental health benefits?',
                  'Do you think that discussing a mental health issue with your employer would have negative consequences?',
                  'Would you be willing to discuss a mental health issue with your coworkers?',
                  'Would you be willing to discuss a mental health issue with your direct supervisor(s)?',
                  'Would you bring up a mental health issue with a potential employer in an interview?']]

data142['year'] = 2014
data142 = data142.rename({'Would you be willing to discuss a mental health issue with your coworkers?': 
                          'Would you feel comfortable discussing a mental health disorder with your coworkers?',
                         'Do you work remotely (outside of an office) at least 50% of the time?':
                         'Do you work remotely?',
                         'If you have a mental health condition, do you feel that it interferes with your work?':
                         'If you have a mental health issue, do you feel that it interferes with your work when NOT being treated effectively?'}, axis=1)

data142
```

```python
data = data142.append(data162, ignore_index=True)
data = data.append(data172, ignore_index=True)
```

```python
# filtering the data
# only US
data1 = data[data['Country'].isin(['United States','United States of America'])]
# gender cleaning
data1['Gender'].str.strip()
data1['Gender'] = data1['Gender'].replace(to_replace=["M", "male", 'dude','Cis Male', 'Male-ish', 'Make', 
                                                      'Male (CIS)', 'maile', 'm', 'Male ', 'male ', 'mail',
                                                     'Male (cis)', 'Male.', 'Cis male', 'Sex is male', 'Dude',
                                                     'cisdude', 'cis man', 'cis-male', 'Cis-male','Man', 'msle',
                                                     'cis hetero male', 'God King of the Valajar','Mail','cis male',
                                                     ], 
                                          value="Male")
data1['Gender'] = data1['Gender'].replace(to_replace=['Female','female', 'Trans-female', 'Cis Female', 'F', 'f',
                                                      'Femake', 'woman','Female ', 'Woman', 'cis-female/femme',
                                                      'Trans woman', 'Female (trans)', 'Female (cis)', 'femail',
                                                      'female ', 'fm', 'Female or Multi-Gender Femme', 'female/woman',
                                                      'Cisgender Female', 'mtf', 'fem', 'human',
                                                      'Female (props for making this a freeform field, though)',
                                                      ' Female','Cis-woman', 'AFAB','F, cisgender',
                                                      'female (cis)', 'My sex is female.','female (cisgender)', 
                                                      'Female (cis) ','cis-Female','cis female', 'F','Female-ish','Transfeminine'], 
                                          value="Female")
data1['Gender'] = data1['Gender'].replace(to_replace=['queer/she/they', 'non-binary','Genderqueer',
                                                      'genderqueer', 'Human', 'Genderfluid', 
                                                      'genderqueer woman','Fluid', 'Male/genderqueer', 
                                                      'Nonbinary', 'Unicorn','Genderflux demi-girl', 
                                                      'Agender', 'uhhhhhhhhh fem genderqueer?','Non-binary', 
                                                      'sometimes', 'Contextual', 'Non binary','Genderqueer demigirl', 
                                                      'nonbinary'], 
                                          value="Queer/Non-binary")
data1['Gender'] = data1['Gender'].replace(to_replace=['Nah', 'p', 'Other','none of your business', 'None'], 
                                          value="Others")

data1['Gender'].unique()
```

```python
data1['Age'] = data1['Age'].replace(to_replace=[-29, -1],value=29)
data1['Age'] = data1['Age'].replace(to_replace=[323],value=32)
```

```python
data1
```

```python
data1 = data1.rename({'Have you been diagnosed with a mental health condition by a medical professional?': 
                      'Diagonosed with a mental health condition'}, axis=1)

data1 = data1.rename({'If so, what condition(s) were you diagnosed with?': 
                      'Mental health conditionn(s)'}, axis=1)
```

```python
data1 = data1.rename({'Would you bring up a mental health issue with a potential employer in an interview?': 
                      'Bring up mental health issues in an interview'}, axis=1)
```

```python
data1 = data1.rename({'Would you be willing to discuss a mental health issue with your direct supervisor(s)?': 
                      'Comfortable discussing mental health issues with direct supervisor(s)'}, axis=1)

data1['Comfortable discussing mental health issues with direct supervisor(s)'] = data1['Comfortable discussing mental health issues with direct supervisor(s)'].replace(
    to_replace=["Some of them"], value="Yes")

data1['Comfortable discussing mental health issues with direct supervisor(s)'].unique()
```

```python
data1 = data1.rename({'Would you feel comfortable discussing a mental health disorder with your coworkers?': 
                      'Comfortable discussing mental health issues with corworkers'}, axis=1)

data1['Comfortable discussing mental health issues with corworkers'] = data1['Comfortable discussing mental health issues with corworkers'].replace(
    to_replace=["Some of them"], value="Yes")

data1['Comfortable discussing mental health issues with corworkers'].unique()
```

```python
data1 = data1.rename({'If you have a mental health issue, do you feel that it interferes with your work when NOT being treated effectively?': 
                      'Mental health issue inteferes with work'}, axis=1)
```

```python
data1 = data1.rename({'Do you think that discussing a mental health issue with your employer would have negative consequences?': 
                      'Negative impacts from discussing mental health issues with employer'}, axis=1)

data1['Negative impacts from discussing mental health issues with employer'] = data1['Negative impacts from discussing mental health issues with employer'].replace(
    to_replace=["Negatively"], value="Yes")

data1['Negative impacts from discussing mental health issues with employer'] = data1['Negative impacts from discussing mental health issues with employer'].replace(
    to_replace=["I'm not sure"], value="Maybe")

data1['Negative impacts from discussing mental health issues with employer'] = data1['Negative impacts from discussing mental health issues with employer'].replace(
    to_replace=['Positively', 'No change'], value="No")

data1['Negative impacts from discussing mental health issues with employer'] = data1['Negative impacts from discussing mental health issues with employer'].replace(
    to_replace=['Not applicable to me'], value="NA")

data1['Negative impacts from discussing mental health issues with employer'].unique()
```

```python
data1 = data1.rename({'Does your employer provide mental health benefits?': 'mental health benefits by employer'}, axis=1)
data1['mental health benefits by employer'] = data1['mental health benefits by employer'].replace(
    to_replace=["I don't know"], value="Don't know")
data1['mental health benefits by employer'] = data1['mental health benefits by employer'].replace(
    to_replace=['Not eligible for coverage / N/A','No, none did'], value="No")
data1['mental health benefits by employer'] = data1['mental health benefits by employer'].replace(
    to_replace=['Some did','Yes, they all did'], value="Yes")

data1['mental health benefits by employer'].unique()
```

```python
data1['Do you work remotely?'] = data1['Do you work remotely?'].replace(to_replace=['Never'], value="No")
data1['Do you work remotely?'] = data1['Do you work remotely?'].replace(to_replace=['Always'], value="Yes")
data1['Do you work remotely?'].unique()
```
